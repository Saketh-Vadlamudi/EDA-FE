```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel('Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.head()
```

```python
final_df.info()
```

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0].astype(int)
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1].astype(int)
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2].astype(int)
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
#In arrival time- time and date is specified.
#but we already have date fields extra. so now we focus on time field only
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0].astype(int)
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1].astype(int)
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df['Dep_Time']=final_df['Dep_Time'].str.split(' ').str[0]
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0].astype(int)
final_df['Dep_min']=final_df['Dep_Time'].str.split(':').str[1].astype(int)
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':2})
```

```python
# we dont need route because we already have no.of stops column
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Duration']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
#final_df[final_df['Duration']=='5m']
#we are deleting this 2 rows because data is not present correctly there.
#As it is only 2 records we are not trying to modify it, but deleting it

final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0].astype(int)
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
# we can use dummies but best approach is to use laberl encoding or one hot encoding as it is train ans test date
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
